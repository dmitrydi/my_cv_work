```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = '0'
train_annotations_dir = '../data/cam_1_hand_markup/to_xml_nearbound/train'
val_annotations_dir = '../data/cam_1_hand_markup/to_xml_nearbound/test'
images_dir = 'D:/data/pics/cam_1_src'
```

```python
from darkflow.net.build import TFNet
```

```python
build_path = 'test_build'
```

```python
weights = os.path.join(build_path, 'bin', 'yolo.weights')
model = os.path.join(build_path, 'cfg', 'yolo-people.cfg')
cfg = os.path.join(build_path, 'cfg')
labels = os.path.join(build_path, 'labels.txt')
```

```python
options=dict()
#options['gpu'] = 0.5
options['model'] = model
options['load'] = weights
options['config'] = cfg
options['labels'] = labels
options['train'] = True
options['dataset'] = images_dir
options['val_dataset'] = images_dir
options['annotation'] = train_annotations_dir
options['val_annotation'] = val_annotations_dir
options['epoch'] = 10
options['gpu'] = 0.8
options["batch"] = 2
options["subdivisions"] = 1
options['backup'] = 'D:/data/models/tpe_retrain_nearbound/ckpt'
options['summary'] = 'D:/data/models/tpe_retrain_nearbound/summary'
options['save'] = 100
options['threshold'] = 0.5
options['lr'] = 1e-4
#options['trainable_layers'] = ['50-convolutional', '52-convolutional']
```

```python
tfnet = TFNet(options)
```

```python
tfnet.FLAGS
```

```python
new_op = {'batch': 4}
```

```python
tfnet.FLAGS.update(new_op)
```

```python
losses, train_losses = tfnet.train()
```

```python
print(losses)
```

```python
metrics_saving_path = "D:\\data\\models\\tpe_retrain\\metrics"
metrics_filename = 'losses.json'
```

```python
import json
with open(os.path.join(metrics_saving_path, metrics_filename), 'w') as outfile:
    json.dump(losses, outfile)
```
