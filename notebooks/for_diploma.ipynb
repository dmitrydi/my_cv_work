```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# данные для графика обучения tpe-модели на deep sort фичах
eer_ds = [13.31, 12.15, 11.71, 11.49, 11.23, 11.14, 11.21, 10.83, 10.71, 11.04, 10.72, 10.68]
epochs_ds = np.linspace(1,12,12, dtype='int')
```

```python
# данные для графика обучения на Inception фичах
eer_inception = [9.15, 8.16, 6.76, 6.60, 6.10, 6.15, 5.89, 5.61, 5.59, 5.32, 5.07, 5.01, 5.08, 5.10, 5.26,
                5.00, 4.92, 4.76, 4.96, 4.85]
epochs_inception = np.linspace(1,20,20, dtype='int')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(epochs_ds, eer_ds, label='DeepSort features')
ax.plot(epochs_inception, eer_inception, label = 'Inception features')
ax.set_xlabel('Epoch')
ax.set_ylabel('EER, %')
fig.suptitle('Кривые обучения TPE-классификатора', fontsize=20)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.legend(prop={'size': 16})
plt.ylim(ymin=0)
plt.ylim(ymax=15)
plt.xticks(epochs_inception)
for item in [ax.xaxis.label, ax.yaxis.label]:
    item.set_fontsize(16)
plt.show()
#plt.savefig(os.path.expanduser('~/d_test/data/models/darkflow_retrain_all_imgs/metrics/pr-curves.jpg'))
```

```python
# обучение детектора
epochs_detector = []
losses_detector = []
data = [{'epoch': 0, 'validation_loss': 240},{'epoch': 1, 'validation_loss': 4.868796396255493}, {'epoch': 2, 'validation_loss': 3.530385680993398}, {'epoch': 3, 'validation_loss': 3.3911765257517499}, {'epoch': 4, 'validation_loss': 3.4082509775956473}, {'epoch': 5, 'validation_loss': 3.4444509545962014}, {'epoch': 6, 'validation_loss': 3.4159366418917974}, {'epoch': 7, 'validation_loss': 3.3174688975016275}, {'epoch': 8, 'validation_loss': 3.3082421898841856}, {'epoch': 9, 'validation_loss': 4.0316208084424341}, {'epoch': 10, 'validation_loss': 3.5968279699484507}]
for d in data:
    epochs_detector.append(d['epoch'])
    losses_detector.append(d['validation_loss'])
```

```python
losses_detector = np.array(losses_detector)
fig, ax = plt.subplots(figsize=(15,10))
ax.set_yscale('log')
ax.plot(epochs_detector, losses_detector, label='YOLO detector loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Validation loss')
ax.grid(True)
fig.suptitle('Кривая обучения детектора', fontsize=20)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.legend(prop={'size': 16})
plt.xticks(epochs_detector)
for item in [ax.xaxis.label, ax.yaxis.label]:
    item.set_fontsize(16)
#plt.grid()
plt.show()
```
